# Hugging Face Base Models

```python
#Selecting the classifier

! pip  install transformers
from transformers import pipeline

#Labels: 0 -> Negative; 1 -> Neutral; 2 -> Positive
classifier = pipeline(model="cardiffnlp/twitter-roberta-base-sentiment")
```

```python
video_link = "https://www.youtube.com/watch?v=VYpAodcdFfA"
```

```python
data=[]
label=[]
wait = WebDriverWait(wd,15)
wd.get(video_link)
for item in range(5): 
    wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
    time.sleep(15)
i=0
for comment in wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#content"))):
    i+=1
    if (i < 4):
      continue
    data.append(comment.text)
    label.append(classifier(comment.text)[0]['label'])
```

```python
import pandas as pd   
df = pd.DataFrame(zip(data, label), columns=['comment', 'label'])
pd.set_option("display.max_colwidth", -1)
df.head(25)
```
